```python
from collections import OrderedDict
import geocoder
import pickle
import operator
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="specify_your_app_name_here")
import time
```

```python
try:
    locations=pickle.load(open("locations.pickle","rb"))
except:
    pass
```

```python
def getGeoCoding(name):
    #g = geocoder.google(name)
    print(name)
    location = geolocator.geocode(name)
    print(name,(location.latitude, location.longitude))
    return (location.latitude, location.longitude)

def getCountryGeo(locations,name):
    if name in ["''","'.'",0]:
        return locations
    else:
        if locations[name]==None:
            locations[name]=getGeoCoding(name)
       
        return locations
    
```
