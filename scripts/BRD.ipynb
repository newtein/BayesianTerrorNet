```python
from collections import OrderedDict
import pickle
import numpy as np
import operator
```

```python
filters=open('filters.txt','r').readline().strip().split('|')
file="pipeFormat.csv"
f=open(file,"r")
headers=f.readline().strip().split('|')
f.close()

def getIndex(s):
    return headers.index(s)

def dumpPickle(name,data):
    pickle.dump(data,open('pickles/{}.pickle'.format(name),'wb'))
    
def loadPickle(name):
    return pickle.load(open('pickles/{}.pickle'.format(name),'rb'))
    
```

```python
def getDetails(filterby):
    data={}
    file="pipeFormat.csv"
    noLabel=['.','']
    f=open(file,"r")
    headers=f.readline().strip().split('|')
   
    nagid={}
    targets={}
    yearIndex=getIndex('year')
    nagcodeIndex=getIndex('nagcode_1')
    nagnameIndex=getIndex('nag_name')
    BRD2Index=getIndex(filterby)

    for l in f:
        row=l.split("|")
        year=int(row[yearIndex])
        nagcode=int(row[nagcodeIndex])
        nagname=row[nagnameIndex]
        brd = row[BRD2Index]
        if nagcode not in data:
            data[nagcode]=0
        if brd!='':
            if len(brd.split('/'))>1:
                brd = brd.split('/')[-1]
                brd = '19'+brd if 45<=int(brd)<=99 else '20'+brd
            else:
                brd = brd.split('-')[0]
        data[nagcode] = brd
    f.close()
    return data


def getDetails2(filterby):
    data={}
    file="pipeFormat.csv"
    noLabel=['.','']
    f=open(file,"r")
    headers=f.readline().strip().split('|')
   
    nagid={}
    targets={}
    yearIndex=getIndex('year')
    nagcodeIndex=getIndex('nagcode_1')
    nagnameIndex=getIndex('nag_name')
    BRD2Index=getIndex(filterby)
    notFound = set()

    for l in f:
        row=l.split("|")
        year=int(row[yearIndex])
        nagcode=int(row[nagcodeIndex])
        nagname=row[nagnameIndex]
        brd = row[BRD2Index]
        if nagcode not in data:
            if brd!='':
                if len(brd.split('/'))>1:
                    brd = brd.split('/')[-1]
                    brd = '19'+brd if 45<=int(brd)<=99 else '20'+brd
                else:
                    brd = brd.split('-')[0]
                temp = int(brd) - int(year)
                if temp>=0:
                    data[nagcode] = temp
                else:
                    notFound.add(nagcode)
    f.close()
    print(notFound)
    return data
```

```python
data = getDetails('startdate2')

yearDied = loadPickle('yearDied')
Deadyear={}
for year in yearDied:
    for nagcode in yearDied[year]: 
        if nagcode not in Deadyear:
            Deadyear[nagcode]=year
        else:
            print(nagcode,year)
            
thirivedAfterBRD={}
for nagcode in Deadyear:
    if data[nagcode]!='':
        temp =Deadyear[nagcode] - int(data[nagcode])
        if temp>=0:
            thirivedAfterBRD[nagcode] = Deadyear[nagcode] - int(data[nagcode])

yearLength = loadPickle('yearLength')
x,y = [],[]
for nagcode in thirivedAfterBRD:
    x.append(thirivedAfterBRD[nagcode])
    y.append(yearLength[nagcode]) 
    
np.corrcoef(x,y)
```
