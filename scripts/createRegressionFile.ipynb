```python
import pickle
from collections import OrderedDict
import numpy as np
```

```python
filters=open('filters.txt','r').readline().strip().split('|')
file="pipeFormat.csv"
f=open(file,"r")
headers=f.readline().strip().split('|')
new_headers = filters
f.close()
```

```python
def dumpPickle(name,data):
    pickle.dump(data,open('pickles/{}.pickle'.format(name),'wb'))
    
def loadPickle(name):
    return pickle.load(open('pickles/{}.pickle'.format(name),'rb'))
  
```

```python
organizationInfo = loadPickle('organizationInfo')
for nagcode in organizationInfo:
    ids = loadPickle('ids')
    objs = loadPickle('objs')
    ideology, objective='',''
    for givenId in ids:
        IdData = loadPickle(givenId)
        if nagcode in IdData:
            ideology = givenId
            break
    for givenobj in objs:
        ObjData = loadPickle(givenobj)
        if nagcode in ObjData:
            objective = givenobj
    
    organizationInfo[nagcode] = [ideology,objective]+organizationInfo[nagcode]
            
```

```python
new_headers = ['Ideology','Objective'] + filters
```

```python
Deadyear = loadPickle('Deadyear')
years = list(range(1945,2011))
yearLength = loadPickle('yearLength')
a_IdeologyProb = loadPickle('archive/a_IdeologyProb')
d_IdeologyProb = loadPickle('archive/d_IdeologyProb')
a_ObjectiveProb = loadPickle('archive/a_ObjectiveProb')
d_ObjectiveProb = loadPickle('archive/d_ObjectiveProb')
brd = loadPickle('BRD_NAG')
for nagcode in organizationInfo:
    ideology,objective = organizationInfo[nagcode][0],organizationInfo[nagcode][1]
    yearDied = Deadyear[nagcode]
    a_ideoScore = a_IdeologyProb[ideology][years.index(yearDied)] if ideology!='' else None
    d_ideoScore = d_IdeologyProb[ideology][years.index(yearDied)] if ideology!='' else None
    
    a_objScore = a_ObjectiveProb[objective][years.index(yearDied)] if objective!='' else None
    d_objScore = d_ObjectiveProb[objective][years.index(yearDied)] if objective!='' else None
    #print(nagcode,a_ideoScore,d_ideoScore,a_objScore,d_objScore)
    year_for_brd = brd.get(nagcode,None)
    organizationInfo[nagcode] = organizationInfo[nagcode] + [a_ideoScore,d_ideoScore,a_objScore,d_objScore,year_for_brd,yearLength[nagcode]]
```

```python
new_headers = new_headers[2:] + ['Ideology (Active Support)','Ideology (De-facto Support)','Objective (Active Support)','Objective (De-facto Support)','Year from BRD','Life-span']
```
